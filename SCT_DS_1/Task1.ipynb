```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Optional: make charts look nicer
plt.style.use("seaborn-v0_8")
```

```python
# Load dataset (make sure the CSV file is in the same folder as this notebook)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_555464.csv", skiprows=4)

# Preview first few rows
df.head()
```

```python
# Select data for the year 2022
year = "2022"
data_2022 = df[["Country Name", year]].dropna()

# Quick check
data_2022.head()
```

```python
# Bar Chart - Top 10 most populated countries in 2022
top10 = data_2022.sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10[year]/1e6, color="skyblue")
plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries in 2022")
plt.ylabel("Population (Millions)")
plt.show()
```

```python
# Histogram - Distribution of populations across all countries (2022)
plt.figure(figsize=(8,6))
plt.hist(data_2022[year]/1e6, bins=30, color="lightgreen", edgecolor="black")
plt.title("Population Distribution Across Countries (2022)")
plt.xlabel("Population (Millions)")
plt.ylabel("Number of Countries")
plt.show()
```
